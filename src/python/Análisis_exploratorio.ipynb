```python
# importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from util.basic import load_arrow
import os
```

# Cargar datos

```python

# obtener los nombres de los archivos que contienen los datos
times = os.listdir('/content/drive/MyDrive/Escuela/data/BTCUSDT_30m/')
# excluimos el summary de los archivos
times.pop(0)
# aseguramos que se cumpla el orden cronologico
times.reverse()
#juntamos todos los datos en un solo data frame
frames = []
for t in times:
  df = pd.read_feather("/content/drive/MyDrive/Escuela/data/BTCUSDT_30m/"+ t)
  frames.append(df)
data = pd.concat(frames)
data
```
